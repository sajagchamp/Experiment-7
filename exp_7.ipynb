```python
#factorial of a number using decrement

n=int(input("enter a number:"))
fact=1
while n>0:
  fact=fact*n
  n=n-1
print("factorial is:",fact)

```

```python
#Reverse of a number

num=int(input("enter a number:"))
rev=0
while num>0:
  digit=num%10
  rev=rev*10+digit
  num=num//10
print("reverse is:",rev)
```

```python
#check palindrome number

num=int(input("enter a number:"))
temporary=num
rev=0
while num >0:
  rev=rev*10+num%10
  num=num//10
if temporary==rev:
  print("the number is palindrome")
else:
  print("the number is not palindrome")
```

```python
#check palindrome string for a purticular string

s="madam"
i,j=0,len(s)-1
is_palindrome=True
while i<j:
  if s[i]!=s[j]:
    is_palindrome=False
    break
  i +=1
  j -=1
if is_palindrome:
  print("the string is palindrome")
else:
  print("the string is not palindrome")
```

```python
#check palindrome string for generalised

s=input("enter a string:")
i,j=0,len(s)-1
is_palindrome=True
while i<j:
  if s[i]!=s[j]:
    is_palindrome=False
    break
  i +=1
  j -=1
if is_palindrome:
  print("the string is palindrome")
else:
  print("the string is not palindrome")
```

```python
#check palindrome number/string

st =input("enter the string:")
rev= st[::-1]
if st==rev:
  print("the string is palindrome")
else:
  print("the string is not palindrome")
```

```python
#search an element in a list

num=[10,20,30,40,50]
key=int(input("enter element to search:"))
i=0
while i<len(num):
  if num[i]==key:
    print("element found at index",i)
    break
  i+=1
else:
  print("element not found")
```
